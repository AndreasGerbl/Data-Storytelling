Reporterstellung (pptx-Erstellung)

```python
from pptx.util import Inches
from pptx import Presentation

X = Presentation()

Layout = X.slide_layouts[0]
first_slide = X.slides.add_slide(Layout)

first_slide.shapes.title.text = "Report" 
first_slide.placeholders[1].text = "deutsche Immobilienmarktanalyse"

Second_Layout = X.slide_layouts[5]
second_slide = X.slides.add_slide(Second_Layout)
second_slide.shapes.title.text = "Immobilienpreisentwicklung"

textbox = second_slide.shapes.add_textbox(Inches(1), Inches(5.3),Inches(8), Inches(4)) 
textframe = textbox.text_frame
paragraph = textframe.add_paragraph()
paragraph.text = "Westliche Großstädte erlebten eine stark positive Preisentwicklung vor allem im Vergleich zu ostdeutschen Städten. \nFokus auf Investitionen in Städten über dem Durchschnitt."
paragraph.level = 1
left = Inches(1.4)
top = Inches(1.85)
pic = second_slide.shapes.add_picture('images/fig.PNG', left, top, Inches(7.3))

Third_Layout = X.slide_layouts[5]
third_slide = X.slides.add_slide(Third_Layout)
third_slide.shapes.title.text = "Ost-West-Vergleich"

textbox = third_slide.shapes.add_textbox(Inches(1), Inches(5.5),Inches(8), Inches(4)) 
textframe = textbox.text_frame
paragraph = textframe.add_paragraph()
paragraph.text = "Starker Unterschied zwischen Ost und West über die letzten 30 Jahre. \nFaktoren wie Finanzkrise erhöhen die Differenz weiter wie auch aktuelle Covid-Pandemie und Wohnungsmangel in Deutschland."
paragraph.level = 1
left = Inches(1.52)
top = Inches(1.85)
pic = third_slide.shapes.add_picture('images/fig2.PNG', left, top, Inches(6.8))

Fourth_Layout = X.slide_layouts[5]
fourth_slide = X.slides.add_slide(Fourth_Layout)
fourth_slide.shapes.title.text = "Modernisierungsbedarf"

textbox = fourth_slide.shapes.add_textbox(Inches(1), Inches(5.5),Inches(8), Inches(4))  
textframe = textbox.text_frame
paragraph = textframe.add_paragraph()
paragraph.text = "Große Investitionen wie der Ausbau des Dachbodens oder Kellers sowie zusätzlicher Räume (Anbau) werden von immer mehr deutschen gewünscht. \nAktuelle Situation lässt Menschen realisieren, was benötigt wird (statt kleinerer Anpassungen lieber ein extra Home Office Raum)."
paragraph.level = 1
left = Inches(1.52)
top = Inches(1.85)
pic = fourth_slide.shapes.add_picture('images/fig1.PNG', left, top, Inches(6.8))

Fifth_Layout = X.slide_layouts[5]
fifth_slide = X.slides.add_slide(Fifth_Layout)
fifth_slide.shapes.title.text = "Wertsteigerung"

textbox = fifth_slide.shapes.add_textbox(Inches(1), Inches(5.5),Inches(8), Inches(4)) 
textframe = textbox.text_frame
paragraph = textframe.add_paragraph()
paragraph.text = "Große Investitionen führen zu hohen Immobilienwertsteigerungen. Risikobereitschaft ist gefragt, um Immobilien an Markt anzupassen und auf Kundenwünsche und -bedürfnisse einzugehen. \nFokus auf Modernisierung statt Restaurierung. "
paragraph.level = 1
left = Inches(1.52)
top = Inches(1.85)
pic = fifth_slide.shapes.add_picture('images/fig7.PNG', left, top, Inches(6.8))

Sixth_Layout = X.slide_layouts[5]
sixth_slide = X.slides.add_slide(Sixth_Layout)
sixth_slide.shapes.title.text = "Saisonaler Markt"

textbox = sixth_slide.shapes.add_textbox(Inches(0.5), Inches(4.2),Inches(9), Inches(4.7)) 
textframe = textbox.text_frame
paragraph = textframe.add_paragraph()
paragraph.text = "Die durchschnittlichen Verkaufspreise steigen in der Regel im Frühjahr und Herbst, während sie im Hochsommer und Winter fallen. \nDie steigenden Verkaufszahlen im Sommer deuten auf erhöhten Bedarf und Interesse hin während die niedrigen Verkäufe im April und Oktober vor allem auf die sehr hohen Preise zurückzuführen sind. \nOptimale Kaufzeitpunkte bei Preiseinbrüchen im Winter. Verkaufszeitpunkte zwischen April und Oktober entweder bei hoher Nachfrage (Hochsommer) oder für maximalen Gewinn im April und Oktober (Vorsicht! Sehr geringe Nachfrage)."
paragraph.level = 1
left = Inches(0.3)
top = Inches(2)
pic = sixth_slide.shapes.add_picture('images/fig3.PNG', left, top, Inches(9.3))

X.save("python_created3.pptx")
```
